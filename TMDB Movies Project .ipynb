12. Classify movies based on ratings[Excellent, Good and Average] and count each ratings

```python
data.columns
    
```

```python
def vote_average(vote_average):
    if vote_average>=7:
        return "Excellent"
    elif vote_average>=6:
        return "Good"
    else:
        return "Average"
```

```python
data["rating_cat"]=data["vote_average"].apply(vote_average)
```

```python
data.head(4)
```

```python
data["rating_cat"].value_counts()
```

```python
data["rating_cat"].value_counts(normalize=True)
```
